```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
city_data_to_load = "data/city_data.csv"
ride_data_to_load = "data/ride_data.csv"

# Read the City and Ride Data
city_data = pd.read_csv(city_data_to_load)
ride_data = pd.read_csv(ride_data_to_load)

# Combine the data into a single dataset
city_ride_data = pd.merge(city_data, ride_data, how="left", on=["city","city"])

# Display the data table for preview
city_ride_data
```

```python
organized_city_ride_data = city_ride_data[["city", "date", "fare", "ride_id", "driver_count", "type"]]
organized_city_ride_data.head()

organized_city_ride_data.groupby("city")
organized_city_ride_data.head()
```

```python
urban = city_ride_data[(city_ride_data["type"] == "Urban")]
rural = city_ride_data[(city_ride_data["type"] == "Rural")]
suburban = city_ride_data[(city_ride_data["type"] == "Suburban")]
print(urban)
print(rural)
print(suburban)
```

```python
urban_avg = urban.groupby("city")["fare"].mean()
rural_avg = rural.groupby("city")["fare"].mean()
suburban_avg = suburban.groupby("city")["fare"].mean()
print(urban_avg)
print(rural_avg)
print(suburban_avg)
```

```python
urban_rides = urban.groupby(["city"]).count()["ride_id"]
rural_rides = rural.groupby(["city"]).count()["ride_id"]
suburban_rides = suburban.groupby(["city"]).count()["ride_id"]
print(urban_rides)
print(rural_rides)
print(suburban_rides)
```

```python
urban_drivers = urban.groupby("city")["driver_count"].mean()
rural_drivers = rural.groupby("city")["driver_count"].mean()
suburban_drivers = suburban.groupby("city")["driver_count"].mean()
print(urban_drivers)
print(rural_drivers)
print(suburban_drivers)
```

## Bubble Plot of Ride Sharing Data

```python
# Obtain the x and y coordinates for each of the three city types
x_axis = [urban_rides, rural_rides, suburban_rides]
y_axis = [urban_avg, rural_avg, suburban_avg]

# Build the scatter plots for each city types
urban_plot = plt.scatter(urban_rides, urban_avg, linewidths=1, marker="o", facecolor="lightcoral", edgecolor="black", s=10*urban_drivers, alpha=0.8, label="Urban")
rural_plot = plt.scatter(rural_rides, rural_avg, linewidths=1, marker="o", facecolor="gold", edgecolor="black", s=7*rural_drivers, alpha=0.8, label="Rural")
suburban_plot = plt.scatter(suburban_rides, suburban_avg, linewidths=1, marker="o", facecolor="lightskyblue", edgecolor="black", s=8*suburban_drivers, alpha=0.8, label="Suburban")


# Incorporate the other graph properties
plt.title("Pyber Ride Sharing (2016)")
plt.xlabel("Total Number of Rides (Per City)")
plt.ylabel("Average Fare ($)")

plt.grid(True)

# Create a legend
lgnd=plt.legend(["Urban", "Rural", "Suburban"], title="City Types", loc='upper right')
lgnd.legendHandles[0]._sizes = [30]
lgnd.legendHandles[1]._sizes = [30]
lgnd.legendHandles[2]._sizes = [30]

# Incorporate a text label regarding circle size
plt.text(42,35,"Note:\n Circle size correlates with driver count per city") 

# Save Figure
plt.show()
```
